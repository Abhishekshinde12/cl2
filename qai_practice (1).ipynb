## Practical 1 - random number

```python
from qiskit import QuantumCircuit
from qiskit.primitives import Sampler
from qiskit_aer import Aer
from qiskit.visualization import circuit_drawer
```

```python
# creating quantum circuit with 16 quibits and 16 classical bits
# classicla bits - used to store mesasurement
# as quantum states collapse into 0/1 upon mesasurement
# classical bits - they allow to measure and store the bits
circuit = QuantumCircuit(16,16)
```

```python
# apply hadamard gate to each of the 16 qubits
# to create superposition state for each qubit
# and yield random number when measured
circuit.h(range(16))
```

```python
# measure each qubit and stores the result in respective classical bits
circuit.measure(range(16), range(16))
```

```python
# printing circuit
print(circuit)
```

```python
# text based visualization of the circuit
circuit_drawer(circuit, output='text')
```

```python
# simulate quantum circuits
simulator = Aer.get_backend('qasm_simulator')
# job - stores the execution of the circuit on the simulator and the shots indicate teh number of times
# the circuit will be executed
job = simulator.run(circuit, shots=1)
# retrieve result of simulation - include measurement and metadata
result = job.result()
# get_counts - returns dictionary of measurement outcomes
# since shots = 1, hence contain one key-value pair
counts = result.get_counts(circuit)
# list(counts.keys())[0] - retrieves single binary string
# int(..,2) - converts this to decimal string
random_number = int(list(counts.keys())[0], 2)

# bin(random_number)[2:] - converts the decimal int back to a binary string excluding 0b prefix
# .zfill(16) - ensure binary string has exactly 16 bits, padding with leading zeros if necessary
binary_number = bin(random_number)[2:].zfill(16)
print("Random number (decimal):", random_number)
print("Random number (binary):", binary_number)
```

## Practical 2 - noise with error correction

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
from qiskit_aer.noise import NoiseModel, depolarizing_error
```

```python
# 3 qubit quantum circuit
qc = QuantumCircuit(3,3)
```

```python
# apply hadamard gate to qubit number 0
qc.h(0)
# creating entangled states between the bits (0,1) and (0,2)
qc.cx(0,1)
qc.cx(0,2)
# measure each qubit, storing the results in corresponding classical bits
qc.measure([0,1,2],[0,1,2])
```

```python
# to simulate realistic quantum hardware errors
noise_model = NoiseModel()
```

```python
# Add depolarizing noise for single-qubit gates, 1% chance of probability
depol_1q = depolarizing_error(0.01, 1)
# Add deploarizing noise for two-qubit gates, 2% chance of probability
depol_2q = depolarizing_error(0.02, 2)

# apply noise to all gates
# apply noise to single qubit gates - u3, x, h
noise_model.add_all_qubit_quantum_error(depol_1q, ['u3', 'x', 'h'])
# apply noise to the CNOT gate
noise_model.add_all_qubit_quantum_error(depol_2q, ['cx'])
```

```python
backend = AerSimulator()
```

```python
# to run circuit with noise, number of shots = 1024
result_with_noise = backend.run(qc, noise_model = noise_model, shots=1024).result()
```

```python
# contains result counts for each outcome with added noise effects
noisy_counts = result_with_noise.get_counts(qc)
```

```python
# assumption that 2% of the errors can be attributed to noise in two-qubit gates, so it reduces each count slightly to partially correct for this.
# hence decreasing the counts by 2% by multiplying by 0.98
# as it simulates real world execution - error occurs
# hence the corrected outcome must lie close to the theoretical values
mitigated_counts = {key: noisy_counts[key] * (1 - 0.02) for key in noisy_counts}
```

```python
# Plot the mitigated results
plot_histogram(mitigated_counts)
```

```python
print("Original counts:")
print(noisy_counts)
```

```python
print("Mitigated counts:")
print(mitigated_counts)
```

```python
plot_histogram([noisy_counts, mitigated_counts], legend=['Original', 'Mitigated'])
```

## Practical 3 - tarrataca quantum production system with 3 puzzle problem

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit.circuit.library import GroverOperator
from qiskit_aer import AerSimulator
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_3_puzzle_circuit():

  # creating quantum circuit
  qr = QuantumRegister(3, 'q')
  cr = ClassicalRegister(3, 'c')
  qc = QuantumCircuit(qr, cr)

  # entangle the (0 and 1 qubits)
  qc.cx(qr[0], qr[1])
  # entangle the (1 and 2 qubits)
  qc.cx(qr[1], qr[2])

  # apply hadamard gates to all
  qc.h(qr[0])
  qc.h(qr[1])
  qc.h(qr[2])

  # separate circuit from the main circuit
  # here we try to set the target that we need to achieve
  # by using this oracle_gate - we phase flip the target
  # which is then used by grover algo to amplify the amplitude of the target
  oracle = QuantumCircuit(3)
  # applying z gate to the first qubit
  oracle.z(0)

  # creating an custom oracle_gate from the oracle circuit
  oracle_gate = oracle.to_gate(label='Oracle')
  # adding a diffusion operator to the oracle gate
  # this operator amplify the probability of the correct solution in the quantum search problem
  grover_operator = GroverOperator(oracle)
  # appending to the grover operator to the main circuit
  qc.append(grover_operator, qr)
  # measures quantum register and stores in the classical register
  qc.measure(qr, cr)

  # return the quantum circuit
  return qc
```

```python
qc = create_3_puzzle_circuit()
simulator = AerSimulator()
# converts the circuit into lower level instructions, minimizing operations and making circuit compatible with the backend
transpiled_qc = transpile(qc, simulator)
# run transpiled code for the defined shots
job = simulator.run(transpiled_qc, shots=1024)
# get the result
result = job.result()
counts = result.get_counts()
print("Measurement results from the 3-puzzle quantum circuit : ")
# the output represents the 3 specific bit binary string that satisfy a certain
# condition, as defined by the oracle. this string is marked state
# the binary string with max occurrences among all the executions indicate the solution to the problem
print(counts)
```

```python
# plotting the keys and the values
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.xlabel('Measurement Outcomes')
plt.ylabel('Counts')
plt.title('3-Puzzle Quantum Circuit Measurement Results')
plt.show()
```

## Practical 4 - teleportation

```python
from qiskit.circuit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
from qiskit import transpile
```

```python
circuit = QuantumCircuit(3,3)
```

```python
# applies an X gate - flip the state of the qubit
circuit.x(0) b
# apply a barrier to the qubit, it doesn't affect the computation, just
# creates a visual separation in the circuit for debugging
circuit.barrier(range(3))
# hadamard gate to qubit 1 - 50:50 probability of measuring probability
circuit.h(1)
# entangle 1 and 2
circuit.cx(1,2)
# entangle 0 and 1
circuit.cx(0,1)
# hadamard gate to qubit 0
circuit.h(0)
# barrier
circuit.barrier(range(3))
# measure first 2 bits
circuit.measure(range(2),range(2))
# create barrier
circuit.barrier(range(3))
# now entangle the 1 and 2
circuit.cx(1,2)
# phase shift of pi
circuit.cz(0,2)
circuit.draw()
```

```python
backend = AerSimulator()
```

```python
# efficient circuit for executing on the backend
qc_compiled = transpile(circuit, backend)
```

```python
# jobs
job_sim = backend.run(qc_compiled, shots=1024)
result_sim = job_sim.result()
```

```python
# return the results of the circuit
counts = result_sim.get_counts(qc_compiled)
print(counts)
```

## Practical 5 - quantum fourier transform

```python
import numpy as np
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit,transpile
from qiskit_aer import AerSimulator
```

```python
from qiskit.circuit.library import QFT
```

```python
# Set up the AerSimulator backend
simulator = AerSimulator()
```

```python
# circuit creation
q = QuantumRegister(5,'q')
c = ClassicalRegister(5,'c')
circuit = QuantumCircuit(q,c)
```

```python
circuit.h(q)
```

```python
# flip states of 0,2,4 qubit
circuit.x(q[4])
circuit.x(q[2])
circuit.x(q[0])
```

```python
# append QFT circuit to original circuit
qft_circuit = QFT(num_qubits=5, approximation_degree=0, do_swaps=True, inverse=False, insert_barriers=False, name="qft")
circuit.append(qft_circuit, q)
```

```python
# compose - mixing
# that is it stacks 2 circuits one on top of another
circuit = circuit.compose(qft_circuit)
# Measure the qubits and store the result in classical register
circuit.measure(q,c)
circuit.draw()
```

```python
# generate efficient code
transpiled_circuit = transpile(circuit, simulator)
```

```python
# run the code
job = simulator.run(transpiled_circuit,shots=1000)
```

```python
print("Job is running...")
print(f"Final job status: {job.status()}")
```

```python
# print the results
job_result = job.result()
counts = job_result.get_counts()
print("\n QFT Output")
print("-------------")
# number of times a bitstring occurs
print(counts)
```
